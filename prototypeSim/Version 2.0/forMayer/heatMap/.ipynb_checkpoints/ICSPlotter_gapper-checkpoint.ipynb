```python
#import statements

import math
import matplotlib as  mpl
import matplotlib.pyplot as plt
import numpy as np
import numpy.random as rande
import time

mpl.rcParams['savefig.dpi'] = 500
mpl.rcParams['figure.dpi'] = 500
```

```python
ics = open("ICSData.txt", "r")
icsStr = ics.read()
icsBigArr = icsStr.split('\n\n')
tmv = []
dpv = []
dv = []
stv = []

for icsArrRun in icsBigArr[:-1]:
        icsArr = icsArrRun.split('\n')
        dpvr = []
        dvr = []
        stvr = []
        
        tmvr = int(float(((icsArr[0]).split('\t'))[1]))
        
        for dat in icsArr[2:-1]:
            splitPair = dat.split('\t')
            dpvr.append(float(splitPair[0]))
            dvr.append(float(splitPair[1]))
            stvr.append(float(splitPair[2]))
        
        tmv.append(tmvr)
        dpv.append(dpvr)
        dv.append(dvr)
        stv.append(stvr)
    
ics.close()

keepr = False

titdv = "Confinment as a Function of Dot Potential"

fileTime = str(int(time.time()))

fileStrdv = "data/confinement_vs_dp(" + fileTime + ")"

fig = plt.figure()
ax = plt.subplot(111)

for itera in range(len(tmv)):
    plt.plot(dpv[itera], dv[itera], label = r"$V_{top}$=" + str(tmv[itera]) + "mV")
plt.grid()

# Shrink current axis by 20%
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])

# Put a legend to the right of the current axis
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.xlabel("Dot Potential (mV)")
plt.ylabel("Dot Confining Voltage (meV/e)")
plt.title(titdv)
if(keepr):
    plt.savefig(fileStrdv+ ".png")
plt.show()


titstv = "Sommer-Tanner as a Function of Dot Potential"

fileStrstv = "data/st_vs_dp(" + fileTime + ")"

fig = plt.figure()
ax = plt.subplot(111)

for itera in range(len(tmv)):
    plt.plot(dpv[itera], stv[itera], label = r"$V_{top}$=" + str(tmv[itera]) + "mV")
plt.grid()

# Shrink current axis by 20%
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])

# Put a legend to the right of the current axis
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.xlabel("Dot Potential (mV)")
plt.ylabel("Sommer Tanner Voltage (meV/e)")
plt.title(titstv)
if(keepr):
    plt.savefig(fileStrstv+ ".png")
plt.show()


fileStr = "data/data(" + fileTime + ")"

if(keepr):
    icsWriter = open(fileStr + ".txt", "w+")
    icsWriter.write(icsStr)
    icsWriter.close()
```

```python
fig = plt.figure()
ax = plt.subplot(111)
slopesc = []
intersc = []

for itera in range(len(tmv)):
    plt.plot(dpv[itera], dv[itera], label = r"$V_{top}$=" + str(tmv[itera]) + "mV")
    slope, intercept = np.polyfit(dpv[itera], dv[itera], 1)
    slopesc.append(slope)
    intersc.append(intercept)
plt.grid()

# Shrink current axis by 20%
box = ax.get_position()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])

# Put a legend to the right of the current axis
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))

plt.xlabel("Dot Potential (mV)")
plt.ylabel("Dot Confining Voltage (meV/e)")
plt.title(titdv)
plt.show()
```
