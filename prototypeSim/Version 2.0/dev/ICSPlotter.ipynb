```python
#import statements

import math
import matplotlib as  mpl
import matplotlib.pyplot as plt
import numpy as np
import numpy.random as rande
import time

mpl.rcParams['savefig.dpi'] = 500
mpl.rcParams['figure.dpi'] = 500
```

```python
ics = open("ICSData.txt", "r")
icsStr = ics.read()
icsArr = icsStr.split('\n')
pos = []
volt = []

for pair in icsArr[3:-1]:
    splitPair = pair.split('\t')
    #print(splitPair)
    pos.append(float(splitPair[0]))
    volt.append(float(splitPair[1]))
    
ics.close()

height = int(float(((icsArr[0]).split('\t'))[1]))
gap = int(float(((icsArr[1]).split('\t'))[1]))
hstr = str(height)
gstr = str(gap)
tit = "Voltage as a Function of Position"
tit += "\n (Height of " + hstr + "nm and Gap of " + gstr + "nm)"

plt.plot(pos, np.multiply(volt, 0.5))
plt.grid()
plt.xlabel("Position (nm)")
plt.ylabel("Voltage (mV)")
plt.title(tit)
#plt.savefig("volt_vs_pos_real.png")
plt.show()
```
