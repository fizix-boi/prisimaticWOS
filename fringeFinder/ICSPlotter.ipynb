```python
#import statements

import math
import matplotlib as  mpl
import matplotlib.pyplot as plt
import numpy as np
import numpy.random as rande
import time

mpl.rcParams['savefig.dpi'] = 500
mpl.rcParams['figure.dpi'] = 500
```

```python
ics = open("ICSData.txt", "r")
icsStr = ics.read()
icsArr = icsStr.split('\n')
pos = []
volt = []

for pair in icsArr[2:-1]:
    splitPair = pair.split('\t')
    pos.append(float(splitPair[0]))
    volt.append(float(splitPair[1]))
    
ics.close()

if(True):
    SMALL_SIZE = 20
    MEDIUM_SIZE = 24
    BIGGER_SIZE = 30
    
    plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
    plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
    plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
    plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
    plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
    plt.rc('legend', fontsize=int(SMALL_SIZE * 0.7))    # legend fontsize

height = int(float(((icsArr[0]).split('\t'))[1]))
hstr = str(height)
tit = "Energy Barrier as a Function of Position"
tit += "\n (Height of " + hstr + "nm)"

fileStr = "data/volt_vs_pos(" + str(int(time.time())) + ")"

volt = np.subtract(volt[0], volt)

plt.plot(pos, volt, linewidth=5)
plt.plot(50 * np.ones(len(volt)), volt, linewidth=3, ls = "--", color = "k")
plt.plot(-50 * np.ones(len(volt)), volt, linewidth=3, ls = "--", color = "k")
#plt.yscale("log")
#plt.xscale("log")
#plt.grid()
plt.xlabel("Position (nm)")
plt.ylabel("Energy (meV)")
#plt.xlim([45e3, 55e3])
#plt.title(tit)

if(False):
    plt.savefig(fileStr + ".png")
    icsWriter = open(fileStr + ".txt", "w+")
    icsWriter.write(icsStr)
    icsWriter.close()

#plt.savefig("barrierPlot_NOGRID.png", bbox_inches = 'tight', pad_inches = 0.1)
plt.show()
```
